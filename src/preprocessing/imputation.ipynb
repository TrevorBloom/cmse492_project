```python
import pandas as pd
import numpy as np
```

```python
X = pd.read_csv("../../data/raw/train.csv")
X.head()
```

```python
X = X.drop(["customer_id", "name", "gender"], axis = 1)

X["owns_house"] = X["owns_house"].map({"Y": 1, "N": 0})
X["owns_car"] = X["owns_car"].map({"Y": 1, "N": 0})

X.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)
#encoude the occupation type feature as this will help with imputation
X["occupation_type"] = encoder.fit_transform(X[["occupation_type"]])

X.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
#choose iterive imputer as I expect this to be a good imputation on more complex models
imputer = IterativeImputer()
X_imputed = imputer.fit_transform(X)

X_imputed = pd.DataFrame(
    X_imputed, 
    columns=X.columns, 
    index=X.index
)

X_imputed.head()
```

```python
missing = X_imputed.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

features = ['credit_limit_used(%)', 'credit_score', 'prev_defaults', 'default_in_last_6months']
target = 'credit_card_default'

x = X_imputed[features]
y = X_imputed[target]

X_train, X_val, y_train, y_val = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_val_pred = model.predict(X_val)
y_val_prob = model.predict_proba(X_val)[:,1]

# Classification report
print(classification_report(y_val, y_val_pred))

# Confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_val, y_val_pred))

# ROC-AUC
roc_auc = roc_auc_score(y_val, y_val_prob)
print(f"ROC-AUC Score: {roc_auc:.3f}")
```

Very small increase in performance of 0.01 for recall and f1 in the defaulting class, while this is small, this is expected for a simple linear model like a logistic regression.
